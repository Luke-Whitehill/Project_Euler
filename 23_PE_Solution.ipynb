```python
import time
import math
start_time = time.time()

def perfectnum(n):
    summ = 0
    for i in range(1,n//2 + 1):
        if n % i == 0:
            summ += i
        if summ == n:
            return summ



print(time.time() - start_time," seconds")
```

```python
start_time = time.time()

def deficientnum(n):
    summ = 0
    for i in range(1,n//2 + 1):
        if n % i == 0:
            summ += i
    if summ <= n:
        return summ 


            
print(time.time() - start_time," seconds")
```

```python
def abundantnum(n):
    summ = 0 
    for i in range(1,n//2 + 1):
        if n % i == 0:
            summ += i 
    if summ >= n:
        return True
    else:
        return False
```

```python
abnum = []
for i in range(12,28123):
    if abundantnum(i) == True:
        abnum.append(i)
print(abnum)
```
